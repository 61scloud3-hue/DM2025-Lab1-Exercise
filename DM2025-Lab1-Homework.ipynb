```python
### >>> Exercise 6
#1. The number of records has changed: X contains all rows, while X_sample only contains 1000 rows.
#2. The order of records is randomized in X_sample.
#3. Not all rows from X are present in X_sample — some have been left out.
```

```python
### >>> **Exercise 7
X_sample = X.sample(n=1000)
X_sample.category_name.value_counts().plot(kind = 'bar',
                                           title = 'Category distribution',
                                           ylim = [0, X_sample.category_name.value_counts().max()+30],
                                           rot = 0, fontsize = 12, figsize = (8,3))
```

```python
### >>> **Exercise 8
import matplotlib.pyplot as plt

# get counts
X_counts = X.category_name.value_counts()
X_sample_counts = X_sample.category_name.value_counts()

# align the categories (same index order)
categories = X_counts.index
X_counts = X_counts[categories]
X_sample_counts = X_sample_counts[categories]

# create side-by-side bar chart
fig, ax = plt.subplots(figsize=(8,4))

bar_width = 0.35
positions = range(len(categories))

ax.bar([p - bar_width/2 for p in positions], X_counts, 
       width=bar_width, label='X (full dataset)')
ax.bar([p + bar_width/2 for p in positions], X_sample_counts, 
       width=bar_width, label='X_sample')

# formatting
ax.set_xticks(positions)
ax.set_xticklabels(categories, rotation=0)
ax.set_ylabel("Count")
ax.set_title("Category Distribution: Full vs Sample")
ax.legend()

plt.show()
```

```python
### **>>> Exercise 9
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
analyze = count_vect.build_analyzer()
analyze(X.text[0])
```

```python
### **>>> Exercise 10
X_counts = count_vect.fit_transform(X.text)

# pick the 5th record
row_index = 4  

# get the indices of nonzero entries in that row
nonzero_indices = X_counts[row_index].nonzero()[1]

# map indices back to words
words_in_doc5 = [count_vect.get_feature_names_out()[i] for i in nonzero_indices]

print("Words in 5th document that appear at least once:", words_in_doc5)
```

```python
### **>>> Exercise 11
import seaborn as sns
import numpy as np

# choose random 100 docs and 100 terms
docs_sample = np.random.choice(X_counts.shape[0], 100, replace=False)
terms_sample = np.random.choice(X_counts.shape[1], 100, replace=False)

# create sub-matrix
sample_matrix = X_counts[docs_sample, :][:, terms_sample].toarray()

# build dataframe for seaborn
df_todraw = pd.DataFrame(sample_matrix, 
                         index=[f"doc_{i}" for i in docs_sample],
                         columns=[count_vect.get_feature_names_out()[j] for j in terms_sample])

plt.figure(figsize=(12, 10))
sns.heatmap(df_todraw, cmap="PuRd", vmin=0, vmax=1)
plt.title("Heatmap of 100 Docs x 100 Terms")
plt.show()

binary_matrix = (X_counts[docs_sample, :][:, terms_sample] > 0).astype(int).toarray()
sns.heatmap(binary_matrix, cmap="Greys", cbar=False)

nonzeros_per_doc = (X_counts > 0).sum(axis=1)
plt.hist(nonzeros_per_doc, bins=50)
plt.title("Distribution of Nonzero Term Counts per Document")
plt.xlabel("Number of Unique Terms per Doc")
plt.ylabel("Frequency")
plt.show()
```

```python
### >>> **Exercise 12
import plotly.express as px

term_frequencies = []
for j in range(0,X_counts.shape[1]):
    term_frequencies.append(sum(X_counts[:,j].toarray()))
term_frequencies = np.asarray(X_counts.sum(axis=0))[0]

# take first 300 terms and their frequencies
terms = count_vect.get_feature_names_out()[:300]
frequencies = term_frequencies[:300]

# make interactive barplot
fig = px.bar(
    x=terms,
    y=frequencies,
    labels={'x': "Terms", 'y': "Frequency"},
    title="Top 300 Term Frequencies"
)

# rotate x-axis labels
fig.update_layout(
    xaxis=dict(tickangle=90),
    height=600,
    width=1200
)

fig.show()
```
